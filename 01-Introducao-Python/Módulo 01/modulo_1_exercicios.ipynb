## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```
# (A)
tkt_medio19 = 320.52
soma_qtde_vendas19 = 3
codigo_a = tkt_medio19 * soma_qtde_vendas19
print(f'"A": {codigo_a}')
```

```
# (B)
svv_20 = 834.47
tkt_medio_20 = 119.21
codigo_b = svv_20 / tkt_medio_20
print(f'"B": {codigo_b:.0f}')
```

```
# (C)
svv_23 = 15378.12
soma_qtde_vendas23 = 5
codigo_c = svv_23 / soma_qtde_vendas23
print(f'"C": {codigo_c:.2f}')
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'roda mundo, roda gigante, roda moinho, roda pião.'

print(cancao.replace(',',';')) # SUBSTITUI UM TEXTO POR OUTRO DENTRO DE UMA STR -> TROCA DE "," POR ";"
print(cancao.endswith('pião.')) # VERIFICA SE A STR TERMINA COM VALOR ESPECIF. E RETORNA TRUE OU FALSE
print(cancao.count('roda')) # CONTA QUANTAS VEZES A SUB.STRING APARECE NA VARIÁVEL CLASS STR
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2.75% e supera expectativas; é a primeira alta em 6 anos.'

# print(noticia.find('2.75%')) Localizando a posição da str p/ extrair
# print(noticia.find('6')) Localizando a posição da str p/ extrair
posicao_selic = float(noticia[12:16])
posicao_ano = int(noticia[62])
selic = posicao_selic
ano = posicao_ano
print(f'VÁRIAVEL SELIC-> {selic}')
print(f'VÁRIAVEL ANO -> {ano}')
```
